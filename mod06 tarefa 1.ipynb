#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
file = pd.read_csv('b1abb33c4f.csv', sep=",")
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
df = file[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
df_mean = df.groupby(['SEXO', 'ESTCIVMAE']).mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 8))
width = 0.4
ind = np.arange(len(df_mean))

p1 = ax.bar(ind, df_mean['APGAR1'], width, color='lightblue')
p2 = ax.bar(ind + width, df_mean['APGAR5'], width, color='skyblue')

ax.set_xticks(ind + width / 2)
ax.set_xticklabels(df_mean['SEXO'].astype(str) + '-' + df_mean['ESTCIVMAE'].astype(str), rotation=30, fontsize='small')

ax.legend((p1[0], p2[0]), ('APGAR1', 'APGAR5'))

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
df = file[['APGAR5', 'CONSULTAS']]
df_mean = df.groupby(['CONSULTAS']).mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=df_mean, ax=ax)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='CONSULTAS', data=file)
sns.countplot(ax = axes[0, 1], x='SEXO', data=file)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=file)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=file)
```
